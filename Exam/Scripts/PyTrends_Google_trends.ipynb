```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=60)
```

```python
kw_list = ["crypto"] # list of keywords to get data 

pytrends.build_payload(kw_list, cat=0, timeframe='all', geo='', gprop='')

pytrends.interest_over_time()

```

```python
data = pytrends.get_historical_interest(kw_list, year_start=2010, month_start=1, day_start=1, hour_start=0, year_end=2022, month_end=12, day_end=31, hour_end=0, cat=0, sleep=0)
```
